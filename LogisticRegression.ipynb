## Titanic Problem Solving using Logistic Regression Model

```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
y_test = pd.read_csv('gender_submission.csv')
```

```python
test['Survived'] = y_test['Survived']
test
```

Concat data train and data test

```python
frame = [train,test]
df = pd.concat(frame)
df = df.reset_index()
df = df.drop('index', axis=1)
df
```

dropping unnecessary columns

```python
df = df.drop(['Name','Ticket','Cabin','PassengerId'], axis=1)
df
```

Convert categorical feature to numeric

```python
df.Sex=df.Sex.map({'female':0, 'male':1})
df.Embarked=df.Embarked.map({'S':0, 'C':1, 'Q':2,'nan':'NaN'})
df
```

Working with missing Value

```python
median_age_men=df[df['Sex']==1]['Age'].mean()
median_age_women=df[df['Sex']==0]['Age'].mean()
df
```

fill missing values

```python
df.loc[(df.Age.isnull()) & (df['Sex']==0),'Age']=median_age_women
df.loc[(df.Age.isnull()) & (df['Sex']==1),'Age']=median_age_men
df
```

```python
df.isnull().sum()
```

### Feature Scaling

```python
df.Age = (df.Age-min(df.Age))/(max(df.Age)-min(df.Age))
df['Fare'] = (df.Fare-min(df.Fare))/(max(df.Fare)-min(df.Fare))
df
```

```python
df.describe()
```

Splitting Data Manually

```python
X_train = df.loc[0:890]
y_train = X_train['Survived']
X_train = X_train.drop(['Survived'], axis=1)

X_test = df.loc[891:1308]
y_test = X_test['Survived']
X_test = X_test.drop(['Survived'], axis=1)
```

handling missing data again

```python
X_test['Fare'] = X_test['Fare'].fillna(X_test['Fare'].mean())
```

```python
X_test.isnull().sum()
```

```python
X_train.isnull().sum()
```

```python
X_train['Embarked'] = X_train['Embarked'].fillna(0)
```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y_train)

from sklearn.metrics import accuracy_score

Y_pred = clf.predict(X_test)
accuracy_score(y_test, Y_pred)
```

```python
result = pd.DataFrame(test['PassengerId'])
result
```

```python
result['Survived'] = Y_pred
result
```

```python
result.to_csv('submission_using_LogisticRegression.csv',index=False)
```
